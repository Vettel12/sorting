Алгоритм следующий (сортируем по возрастанию):

На каждом шаге проходим по массиву поочередно сравниваем пары соседних элементов.

Если элемент на позиции і больше элемента на позиции i+1, меняем их местами. После первой итерации самый большой элемент окажется в конце массива.

Проходим по массиву, выполняя указанные действия n - 1 раз, или до тех пор, пока на очередной итерации не окажется, что обмены больше не нужны, то есть массив уже отсортирован.

```python
def partition(array, l, r):
    pivot = array[r]
    i = l - 1
    for j in range(l, r):
        if array[j] <= pivot:
            i += 1
            array[i], array[j] = array[j], array[i]
    array[i+1], array[r] = array[r], array[i+1]
    return i+1

def quick_sort(n, array):
    l, r = 0, n-1
    stack = [(l, r)]
    while stack:
        l, r = stack.pop()
        if l < r:
            p = partition(array, l, r)
            stack.append((l, p-1))
            stack.append((p+1, r))
    return array

def binary_search(array, target):
    low = 0
    high = len(array) - 1
    while low <= high:
        mid = (low + high) // 2
        if array[mid] < target:
            low = mid + 1
        elif array[mid] > target:
            high = mid - 1
        else:
            return True
    return False

def solution(n, m, array_first, array_second):
    result = []
    new_array_second = quick_sort(m, array_second)
    new_array_first = set(array_first)
    for i in new_array_first:
        if binary_search(new_array_second, i):
            result.append(i)
    return result[::-1]

def main():
    try:
        n = int(input())
        if n > 10000:
            raise ValueError
        m = int(input())
        if m > 10000:
            raise ValueError
        array_first = list(map(int, input().split()))
        for i in range(n):
            if abs(array_first[i]) > 10000:
                raise ValueError
        array_second = list(map(int, input().split()))
        for i in range(m):
            if abs(array_second[i]) > 10000:
                raise ValueError
        result = solution(n, m, array_first, array_second)
        print(*result)
    except ValueError:
        print("Error: Invalid input")

if __name__ == '__main__':
    main()
```

```python
def partition(array, l, r):
    pivot = array[r]
    i = l - 1
    for j in range(l, r):
        if array[j] <= pivot:
            i += 1
            array[i], array[j] = array[j], array[i]
    array[i+1], array[r] = array[r], array[i+1]
    return i+1

def quick_sort(array, l, r):
    if l < r:
        p = partition(array, l, r)
        quick_sort(array, l, p-1)
        quick_sort(array, p+1, r)
    return array

def binary_search(array, target):
    if not array:
        return 0
    low = 0
    high = len(array) - 1
    count = 0
    while low <= high:
        mid = (low + high) // 2
        if array[mid] < target:
            low = mid + 1
        elif array[mid] > target:
            high = mid - 1
        else:
            count += 1
            # ищем все элементы, равные target
            left = mid - 1
            while left >= 0 and array[left] == target:
                count += 1
                left -= 1
            right = mid + 1
            while right < len(array) and array[right] == target:
                count += 1
                right += 1
            return count
    return 0

def solution(n, m, array_first, array_second):
    result = []
    new_array_second = quick_sort(array_second, 0, m-1)
    for i in array_first:
        count_second = binary_search(new_array_second, i)
        if count_second > 0 and i not in result:
            count_first = array_first.count(i)
            result.extend([i] * min(count_first, count_second))
    return result

def main():
    try:
        n = int(input())
        if n > 10000:
            raise ValueError
        m = int(input())
        if m > 10000:
            raise ValueError
        array_first = list(map(int, input().split()))
        for i in range(n):
            if abs(array_first[i]) > 10000:
                raise ValueError
        array_second = list(map(int, input().split()))
        for i in range(m):
            if abs(array_second[i]) > 10000:
                raise ValueError
        result = solution(n, m, array_first, array_second)
        print(*result)
    except ValueError:
        print("Error: Invalid input")

if __name__ == '__main__':
    main()
```

```python
def merge_sort_iterative(array):
    if len(array) <= 1:
        return array

    n = len(array)
    width = 1

    while width < n:
        for i in range(0, n, 2 * width):
            left = i
            mid = min(i + width, n)
            right = min(i + 2 * width, n)

            # Слияние подмассивов на месте
            merge(array, left, mid, right)

        width *= 2

    return array

def merge(array, left, mid, right):
    # Создаем временный массив для хранения отсортированных элементов
    temp = []
    i = left
    j = mid

    # Слияние двух подмассивов
    while i < mid and j < right:
        if array[i] < array[j]:
            temp.append(array[i])
            i += 1
        else:
            temp.append(array[j])
            j += 1

    # Добавляем оставшиеся элементы из левой половины
    while i < mid:
        temp.append(array[i])
        i += 1

    # Добавляем оставшиеся элементы из правой половины
    while j < right:
        temp.append(array[j])
        j += 1

    # Копируем отсортированные элементы обратно в оригинальный массив
    for k in range(len(temp)):
        array[left + k] = temp[k]

def solution(n, array):
    array_sort = merge_sort_iterative(array)
    for i in range(n-1, 1, -1):
        chunk = array_sort[i-2:i+1]
        if chunk[2] < chunk[0] + chunk[1]:
            break
    return chunk[0] + chunk[1] + chunk[2]

def main():
    try:
        n = int(input())
        if n > 1000:
            raise ValueError
        array = list(map(int, input().split()))
        for i in range(n):
            if abs(array[i]) > 10000:
                raise ValueError
        result = solution(n, array)
        print(result)
    except ValueError:
        print("Error: Invalid input")

if __name__ == '__main__':
    main()
```

Кондратий ввел новый метод сортировки - Относительная сортировка.
Идея метода следующая. С помощью образца - массива уникальных чисел,
задается порядок. В соответствии с этим порядком и должны сортироваться числа.
Но метод еще требует доработки. Пока не известно, как сортировать числа, которые
не входят в образец. Так что такие числа нужно выводить в конце в
соответствии со стандартной сортировкой.

Формат ввода
В первой строке задано количество чисел, которые нужно отсортировать, n.
Оно не превосходит 1000. В следующей строке через пробел заданы n чисел,
каждое из которых не превосходит 1000. В третей строке задана длина образца m.
Это число не превосходит 1000. В следующей строке записан образец. Он состоит из чисел,
не превосходящих 1000. Гарантируется, что среди чисел, которые нужно
отсортировать, присутствуют все числа из образца.

Формат вывода
Выведите в строку через пробел числа, отсортированные Относительной сортировкой.

Ввод:

2 4 3 5 6 0 9 8 7 7

Вывод:

Ввод:

2 3 1 3 2 4 6 7 9 2 19

2 1 4 3 9 6

Вывод:

Ввод:

3 10 5 9 2 7 6 0 8 3 4

8

3 10 5 9 2 7 6 0

Вывод:

3 3 10 5 9 2 7 6 0 4 8

```python
def merge_sort_iterative(array):
    if len(array) <= 1:
        return array

    n = len(array)
    width = 1

    while width < n:
        for i in range(0, n, 2 * width):
            left = i
            mid = min(i + width, n)
            right = min(i + 2 * width, n)

            # Слияние подмассивов на месте
            merge(array, left, mid, right)

        width *= 2

    return array

def merge(array, left, mid, right):
    # Создаем временный массив для хранения отсортированных элементов
    temp = []
    i = left
    j = mid

    # Слияние двух подмассивов
    while i < mid and j < right:
        if array[i] < array[j]:
            temp.append(array[i])
            i += 1
        else:
            temp.append(array[j])
            j += 1

    # Добавляем оставшиеся элементы из левой половины
    while i < mid:
        temp.append(array[i])
        i += 1

    # Добавляем оставшиеся элементы из правой половины
    while j < right:
        temp.append(array[j])
        j += 1

    # Копируем отсортированные элементы обратно в оригинальный массив
    for k in range(len(temp)):
        array[left + k] = temp[k]

def solution(n, array_first, m, array_second):
    counter = {}
    for i in range(n):
        if array_first[i] not in counter:
            counter[array_first[i]] = 1
        else:
            counter[array_first[i]] += 1
    
    result = []
    sec_result = []
    for i in range(m):
        if array_second[i] in counter.keys():
            count = counter[array_second[i]]
            for j in range(count):
                result.append(array_second[i])
                
    for i in range(n):
        if array_first[i] not in array_second:
            sec_result.append(array_first[i])
    return result + merge_sort_iterative(sec_result)   
    
def main():
    try:
        n = int(input())
        if n > 1000:
            raise ValueError
        array_first = list(map(int, input().split()))
        for i in range(n):
            if abs(array_first[i]) > 1000:
                raise ValueError
        m = int(input())
        if m > 1000:
            raise ValueError
        array_second = list(map(int, input().split()))
        for i in range(m):
            if abs(array_second[i]) > 1000:
                raise ValueError
        result = solution(n, array_first, m, array_second)
        print(*result)
    except ValueError:
        print("Error: Invalid input")

if __name__ == '__main__':
    main()
```

Частичная сортировка

После того, как Гоша узнал про сортировку слиянием и быструю сортировку,
он решил придумать свой метод сортировки, который предполагал бы разделение данных на части.
Назвал он свою сортировку Частичной. Этим методом можно отсортировать n уникальных чисел,
находящихся в диапазоне от 0 до n - 1. В соответствии с методом нужно разбить данные на
максимально возможное количество частей таким образом, чтобы можно было отсортировать
каждую из частей отдельно, соединить, и при этом результат будет отсортирован.
После сортировки отдельных частей менять части местами нельзя.

Формат ввода

В первой строке задано n - количество чисел для сортировки.
В следующей строке записан числа от 0 до n - 1, которые нужно отсортировать описанным методом.

Формат вывода

Выведите максимальное количество частей, на которое можно разбить данные при использовании метода Частичной сортировки.

Ввод:

4

0 1 3 2

Вывод:

3

Ввод:

8

3 6 7 4 1 5 0 2

Вывод:

1

Ввод:

5

1 0 2 3 4

Вывод:

4

```python
def solution(n, array):
    current_max = -1
    parts = 0

    for i in range(n):
        current_max = max(current_max, array[i])
        if current_max == i:
            parts += 1

    return parts

def main():
    try:
        n = int(input())
        array = list(map(int, input().split()))
        print(solution(n, array))
    except ValueError:
        print("Error: Invalid input")

if __name__ == '__main__':
    main()
```

Евлампия стала подозревать, что Кондратий отец пытается читать её сообщения и письма.
Она приказала разработать новую систему шифрования. Дано зашифрованное слово и список слов.
Для того, чтобы узнать, какое слово зашифровано, нужно в списке найти самое длинное слово,
которое является подсловом (подпоследовательностью) зашифрованного слова. Если таких слов
более одного, нужно вывести лексикографически минимальное из них.

Формат ввода

В первой строке записано зашифрованное слово. Оно состоит из строчных латинских букв.
Его длина не превосходит 1000 символов. В следующей строке задана дина словаря n,
она не превосходит 1000. В следующих n строках записаны слова из словаря.
Длина каждого из них не превосходит 1000.

Формат вывода

Нужно вывести зашифрованное слово, или пустую строку, если такого слова нет в словаре.

Ввод:

Вывод:

Ввод:

Вывод:

```python
# Функция для проверки, является ли слово подпоследовательностью зашифрованного слова
def is_subsequence(word, encrypted_word):
    i, j = 0, 0
    while i < len(word) and j < len(encrypted_word):
        if word[i] == encrypted_word[j]:
            i += 1
        j += 1
    return i == len(word)

# Функция поразрядной сортировки
def radix_sort_by_length_and_lex(dictionary):
    max_len = max(len(word) for word in dictionary)
    
    # Сортируем по каждому разряду, начиная с конца (для лексикографической сортировки)
    for pos in range(max_len - 1, -1, -1):
        # Counting sort по позиции символа (или по "пустому" символу для коротких слов)
        count = [[] for _ in range(27)]  # 26 букв + 1 для коротких слов
        for word in dictionary:
            if pos < len(word):
                index = ord(word[pos]) - ord('a')  # Символ в слове
            else:
                index = -1  # Пустой символ для коротких слов
            count[index + 1].append(word)
        
        # Обновляем список слов в отсортированном порядке
        dictionary = [word for group in count for word in group]

    # Сортируем по длине, используя поразрядную сортировку (от самой длинной к короткой)
    max_len = max(len(word) for word in dictionary)
    count = [[] for _ in range(max_len + 1)]
    for word in dictionary:
        count[len(word)].append(word)
    
    # Собираем слова в порядке убывания длины
    sorted_dict = [word for group in reversed(count) for word in group]
    return sorted_dict

# Основная функция для поиска самого длинного слова
def solution(word, n, array):
    # Сортируем слова по длине и лексикографически
    sorted_dict = radix_sort_by_length_and_lex(array)
    
    # Ищем первое слово, которое является подпоследовательностью зашифрованного слова
    for candidate in sorted_dict:
        if is_subsequence(candidate, word):
            return candidate
    return ""  # Если нет подходящего слова

def main():
    try:
        word = input()
        if len(word) > 1000:
            raise ValueError
        n = int(input())
        if n > 1000:
            raise ValueError
        array = []
        for i in range(n):
            row = input()
            array.append(row)
        for i in range(n):
            if len(array[i]) > 1000:
                raise ValueError
        # Вызываем решение и выводим результат
        print(solution(word, n, array))
    except ValueError:  
        print("Error: Invalid input")

if __name__ == '__main__':
    main()
```

Формат ввода

Во второй строке записано m - число столбцов матрицы.

неотрицательных целых чисел, каждое из которых не больше 1000.

Формат вывода

```python
def radix_sorted(array):
    length = max(len(str(x)) for x in array)  # Определяем максимальную длину числа в строковом формате
    rang = 10  # Количество "карманов" для цифр от 0 до 9

    for i in range(length):
        count = [[] for _ in range(rang)]

        for x in array:
            figure = x // 10**i % 10 
            count[figure].append(x)

        array = []

        for y in range(rang):
            array.extend(count[y])  # Добавляем все элементы карманов к массиву

    return array
            
def solution(n, m, array):
    # Создаем словарь для хранения элементов каждой диагонали
    diagonals = {}

    # Заполняем словарь элементами матрицы
    for i in range(n):
        for j in range(m):
            k = i - j
            if k not in diagonals:
                diagonals[k] = []
            diagonals[k].append(array[i][j])

    # Сортируем каждую диагональ
    for k in diagonals:
        diagonals[k] = radix_sorted(diagonals[k])

    # Заполняем матрицу отсортированными диагоналями
    for i in range(n):
        for j in range(m):
            array[i][j] = diagonals[i - j].pop(0)
    
    return array

def main():
    try:
        n = int(input().strip())
        if n > 1000:
            raise ValueError
        m = int(input().strip())
        if m > 1000:
            raise ValueError
        
        array = []
        for i in range(n):
            row = list(map(int, input().strip().split()))  # Преобразуем строки в списки чисел
            if len(row) != m:
                raise ValueError
            array.append(row)
        
        # Вызываем решение и выводим результат
        sorted_matrix = solution(n, m, array)
        for row in sorted_matrix:
            print(" ".join(map(str, row)))  # Форматированный вывод
        
    except ValueError:
        print("Error: Invalid input")

if __name__ == '__main__':
    main()
```
